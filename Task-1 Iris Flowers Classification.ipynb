## Label encoder

```python
from sklearn.preprocessing import LabelEncoder
la=LabelEncoder() 
```

```python
data['Species']=la.fit_transform(data['Species'])    #fit_transform: Fit label encoder and return encoded labels.
data.head()
```

```python
X = data.drop(columns=['Species']) 
Y = data['Species']
X[:5] # return list from beginning until index 5.
```

## Spliting the data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y)
```

## Using the supervised ml model


```python
from sklearn.linear_model import LogisticRegression
logi_reg=LogisticRegression()
logi_reg.fit(X_train,Y_train)
```

```python
logi_reg.score(X,Y)
```

```python
logi_reg.score(X_train,Y_train)
```

```python
logi_reg.score(X_test,Y_test)
```
